````python
def filter_text(text:str):
    try:
        if type(text) != str:
            return ''
        else:
            if 'abstract' not in text:
                return text
            else:
                text = text.replace('abstract ', '')
                return text
    except:
        print(f'Error text: {text}...{type(text)}')
````

````python
df['abstract_updated'] = df['abstract'].apply(lambda row: filter_text(row))
````

````python
test_abstracts = df['abstract_updated'][:5]
````

````python
test_abstracts
````

````python
extracted_entities_base = []
````

````python
def perform_ner(text:str) -> list:
    doc = nlp(text)
    entities = [entity for entity in doc.ents]
    return entities
````

````python
_ = test_abstracts.apply(lambda row: extracted_entities_base.extend(perform_ner(row)))
````
