````python
def preprocess_text(text: str):
  '''
  The aim of this function is to preprocess text by converting it to lowercase, and punctuations.
  '''
  text = text.lower()
  text = re.sub(r",\s?",' ', text)
  text = re.sub(r'\.\s*', ' ', text)
  return text
````

````python
def filter_text(text:str):
    try:
        if type(text) != str:
            return ''
        else:
            if 'abstract' not in text:
                return text
            else:
                text = text.replace('abstract ', '')
                return text
    except:
        print(f'Error text: {text}...{type(text)}')
````

````python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples, truncation=True, is_split_into_words=False)

    # Align labels with subword tokens
    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_id = None
        label_ids = []
        for word_id in word_ids:
            if word_id is None:
                label_ids.append(-100)  # Special tokens
            elif word_id != previous_word_id:
                label_ids.append(label[word_id])  # Assign label
            else:
                label_ids.append(-100)  # Subword tokens
            previous_word_id = word_id
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs
````


The goal is to loop through each row, identify the number of sentences, and triplets for each sentence.


````python
def preprocess_scierc_gpt(row):
    sentences = row["sentences"]
    ner = row["ner"]
    relations = row["relations"]

    triplets = []

    for sentence, sentence_ner, sentence_relations in zip(sentences, ner, relations):
        # Convert indices to entity text spans
        entities = {}
        for start, end, _ in sentence_ner:
            entity_text = " ".join(sentence[start:end + 1])
            entities[(start, end)] = entity_text

        # Extract triplets based on relations
        for rel in sentence_relations:
            start1, end1, start2, end2, rel_type = rel
            entity1 = entities.get((start1, end1))
            entity2 = entities.get((start2, end2))
            if entity1 and entity2:
                triplets.append((entity1, rel_type, entity2))

    return triplets
````

````python
scierc_train_df["triplets"] = scierc_train_df.apply(preprocess_scierc_gpt, axis=1)
````

````python
def prepare_training_data(row):
    sentences = row["sentences"]
    triplets = row["triplets"]

    examples = []
    for sentence, triplet in zip(sentences, triplets):
        #input_text = " ".join(sentence)
        input_text = f"Extract entities and relationships from: {' '.join(sentence)}"
        entity1, relation, entity2 = triplet

        # Concatenate entities with the sentence for training
        output_text = f"<entity1> {entity1} </entity1> <relation> {relation} </relation> <entity2> {entity2} </entity2>"
        examples.append((input_text, output_text))

    return examples
````

##### Coreference Resolution:
Trying with fastcoref

https://medium.com/@michaelwood33311/creating-accurate-ai-coreference-resolution-with-fastcoref-20f06044bdf9

````python
from spacy.tokens import Doc, Span
from fastcoref import FCoref
````

````python
def core_logic_part(document: Doc, coref: list[int], resolved: list[str], mention_span: Span):
    final_token = document[coref[1]]
    if final_token.tag_ in ["PRP$", "POS"]:
        resolved[coref[0]] = mention_span.text + "'s" + final_token.whitespace_
    else:
        resolved[coref[0]] = mention_span.text + final_token.whitespace_
    for i in range(coref[0] + 1, coref[1] + 1):
        resolved[i] = ""
    return resolved
````

````python
def get_span_noun_indices(doc: Doc, cluster: list[list[int]]) -> list[int]:
    try:
      spans = [doc[span[0]:span[1]+1] for span in cluster]
      spans_pos = [[token.pos_ for token in span] for span in spans]
      span_noun_indices = [i for i, span_pos in enumerate(spans_pos)
          if any(pos in span_pos for pos in ['NOUN', 'PROPN'])]
      return span_noun_indices
    except Exception as e:
      print(f'Error received while getting span noun indices. {e}. Returning empty list')
      return []
````

````python
def get_cluster_head(doc: Doc, cluster: list[list[int]], noun_indices: list[int]):
    head_idx = noun_indices[0]
    head_start, head_end = cluster[head_idx]
    head_span = doc[head_start:head_end+1]
    return head_span, [head_start, head_end]
````

````python
def is_containing_other_spans(span: list[int], all_spans: list[list[int]]):
    return any([s[0] >= span[0] and s[1] <= span[1] and s != span for s in all_spans])
````

````python
def improved_replace_corefs(document, clusters):
    resolved = list(tok.text_with_ws for tok in document)
    all_spans = [span for cluster in clusters for span in cluster]  # flattened list of all spans

    for cluster in clusters:
        noun_indices = get_span_noun_indices(document, cluster)

        if noun_indices:
            mention_span, mention = get_cluster_head(document, cluster, noun_indices)

            for coref in cluster:
                if coref != mention and not is_containing_other_spans(coref, all_spans):
                    core_logic_part(document, coref, resolved, mention_span)

    return "".join(resolved)
````

````python
def get_fast_cluster_spans(doc, clusters):
    fast_clusters = []
    for cluster in clusters:
        new_group = []
        for tuple in cluster:
            #print(type(tuple), tuple)
            (start, end) = tuple
            #print("start, end", start, end)
            span = doc.char_span(start, end)
            #print('span', span.start, span.end)
            new_group.append([span.start, span.end-1])
        fast_clusters.append(new_group)
    return fast_clusters
````

````python
def get_fastcoref_clusters(doc, text, fcoref_model):
  try:
    preds = fcoref_model.predict(texts=[text])
    fast_clusters = preds[0].get_clusters(as_strings=False)
    fast_cluster_spans = get_fast_cluster_spans(doc, fast_clusters)
  except Exception as e:
    print(f'Error received while getting coreference clusters. {e}. Returning original text')
    return text
  return fast_cluster_spans
````

````python
def transform_biored_to_rebel_format_updated(dataset):
    """
    Transforms the BioRED dataset into REBEL-compatible format with consolidated triplets.

    Args:
        dataset: The BioRED dataset as a list of entries.

    Returns:
        List of dictionaries in the REBEL format.
    """
    rebel_data = []

    for entry in dataset:
        # Concatenate title and abstract
        title = entry['passages'][0]['text'][0]  # Title
        abstract = entry['passages'][1]['text'][0]  # Abstract
        input_text = f"{title} {abstract}"

        # Map entity IDs to entity text
        entity_map = {entity['concept_id']: entity['text'][0] for entity in entry['entities']}

        # Generate triplets and group by head entity
        triplet_groups = {}
        for relation in entry['relations']:
            concept_1 = relation['concept_1']
            concept_2 = relation['concept_2']
            rel_type = relation['type']

            # Ensure both entities exist
            if concept_1 in entity_map and concept_2 in entity_map:
                head = entity_map[concept_1]
                tail = entity_map[concept_2]
                relation_triplet = f"{tail} <obj> {rel_type}"

                # Group relations by head entity
                if head not in triplet_groups:
                    triplet_groups[head] = []
                triplet_groups[head].append(relation_triplet)

        # Consolidate triplets for each head entity
        consolidated_triplets = []
        for head, relations in triplet_groups.items():
            relation_strings = " <subj> ".join(relations)
            consolidated_triplets.append(f"<triplet> {head} <subj> {relation_strings}")

        # Only include rows with valid triplets
        if consolidated_triplets:
            rebel_data.append({"input": input_text, "output": " ".join(consolidated_triplets)})

    return rebel_data
````

````python
def perform_coreference_resolution(text):
    doc = nlp_scispacy(text)
    clusters = get_fastcoref_clusters(doc, text, fcoref_model)
    coref_text = improved_replace_corefs(doc, clusters)
    return coref_text
````

````python
def extract_triplets(output_string):
    """
    Extract triplets from the model's output string.

    Args:
        output_string (str): The raw string output from the REBEL model.

    Returns:
        list of tuples: A list of extracted triplets as (subject, predicate, object).
    """
    pattern = r"<triplet>\s*(.*?)\s*<subj>\s*(.*?)\s*<obj>\s*(.*?)"
    matches = re.findall(pattern, output_string)
    triplets = []
    for match in matches:
        subject, predicate, obj = match
        triplets.append((subject.strip(), predicate.strip(), obj.strip()))
    return triplets
````

````python
def compute_metrics_v1(eval_preds):
    """
    Compute metrics for triplet extraction.

    Args:
        eval_preds (tuple): A tuple (predictions, labels) where:
            predictions (list of str): The model's predicted outputs.
            labels (list of str): The ground truth outputs.

    Returns:
        dict: Precision, Recall, and F1-Score.
    """
    predictions, labels = eval_preds

    # Initialize counts
    tp = 0  # True positives
    fp = 0  # False positives
    fn = 0  # False negatives

    for pred, label in zip(predictions, labels):
        # Extract triplets from predictions and labels
        predicted_triplets = set(extract_triplets(pred))
        ground_truth_triplets = set(extract_triplets(label))

        # Calculate TP, FP, FN
        tp += len(predicted_triplets & ground_truth_triplets)
        fp += len(predicted_triplets - ground_truth_triplets)
        fn += len(ground_truth_triplets - predicted_triplets)

    # Compute Precision, Recall, and F1
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

    return {
        "precision": precision,
        "recall": recall,
        "f1": f1
    }
````

````python
def compute_metrics(eval_preds):
    """
    Compute Precision, Recall, and F1-Score for triplet extraction.
    """
    # Unpack predictions and labels
    predictions, labels = eval_preds

    # Decode predictions and labels into strings
    decoded_predictions = [
        rebel_tokenizer.decode(pred, skip_special_tokens=True) for pred in predictions
    ]
    decoded_labels = [
        rebel_tokenizer.decode(label, skip_special_tokens=True) for label in labels
    ]

    tp, fp, fn = 0, 0, 0

    for pred, label in zip(decoded_predictions, decoded_labels):
        # Extract triplets from predictions and labels
        predicted_triplets = set(extract_triplets(pred))
        ground_truth_triplets = set(extract_triplets(label))

        # Compute True Positives, False Positives, and False Negatives
        tp += len(predicted_triplets & ground_truth_triplets)
        fp += len(predicted_triplets - ground_truth_triplets)
        fn += len(ground_truth_triplets - predicted_triplets)

    # Calculate Precision, Recall, and F1
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

    return {"precision": precision, "recall": recall, "f1": f1}
````

````python
def group_triplets_by_head(triplets):
    """
    Groups triplets by head entity for consolidated output.

    Args:
        triplets (list): List of triplets in the format ['<triplet> Head <subj> Tail <obj> Relation'].

    Returns:
        str: A single consolidated string of grouped triplets.
    """
    grouped_triplets = {}

    for triplet in triplets:
        # Extract Head, Tail, Relation from the triplet
        parts = triplet.replace("<triplet>", "").split("<subj>")
        if len(parts) == 2:
            head = parts[0].strip()
            tail_relation = parts[1].split("<obj>")
            if len(tail_relation) == 2:
                tail = tail_relation[0].strip()
                relation = tail_relation[1].strip()

                # Group by Head entity
                if head not in grouped_triplets:
                    grouped_triplets[head] = []
                grouped_triplets[head].append((tail, relation))

    # Format consolidated output
    formatted_output = []
    for head, relations in grouped_triplets.items():
        relation_strings = " <subj> ".join([f"{tail} <obj> {relation}" for tail, relation in relations])
        formatted_output.append(f"<triplet> {head} <subj> {relation_strings}")

    return " ".join(formatted_output)
````

````python
def transform_biored_to_rebel_format(dataset):
    rebel_data = []

    for entry in dataset:
        # Concatenate title and abstract
        title = entry['passages'][0]['text'][0]  # Title
        abstract = entry['passages'][1]['text'][0]  # Abstract
        input_text = f"{title} {abstract}"

        # Map entity IDs to entity text
        entity_map = {entity['concept_id']: entity['text'][0] for entity in entry['entities']}

        # Generate triplets
        triplets = []
        for relation in entry['relations']:
            concept_1 = relation['concept_1']
            concept_2 = relation['concept_2']
            rel_type = relation['type']

            # Ensure both entities exist
            if concept_1 in entity_map and concept_2 in entity_map:
                triplet = f"<triplet> {entity_map[concept_1]} <subj> {entity_map[concept_2]} <obj> {rel_type}"
                triplets.append(triplet)

        # Only include rows with valid triplets
        if triplets:
            rebel_data.append({"input": input_text, "output": triplets})

    return rebel_data
````

````python
def extract_coref_clusters(tokens, predicted_labels):
    clusters = {}
    token_map = []
    for idx, (token, label) in enumerate(zip(tokens, predicted_labels)):
        if label > 0:  # 0 typically means "no cluster"
            if label not in clusters:
                clusters[label] = []
            # Handle special tokens (e.g., subword tokens starting with ##)
            if token.startswith("##"):
                token = token[2:]
                token_map[-1] += token  # Append subword to previous token
            else:
                clusters[label].append((token, idx))
                token_map.append(token)
        else:
            token_map.append(token)
    return clusters, token_map
````

````python
coref_clusters = extract_coref_clusters(tokens, predicted_labels)
````

````python
def resolve_coreferences(original_text, token_map, clusters):
    resolved_text = original_text
    for cluster_id, mentions in clusters.items():
        if len(mentions) > 1:  # Only process clusters with more than one mention
            antecedent = mentions[0][0]  # First mention is the antecedent
            for mention, idx in mentions[1:]:
                # Replace subsequent mentions in the text with the antecedent
                if mention in resolved_text:
                    resolved_text = resolved_text.replace(mention, antecedent)
    return resolved_text
````
