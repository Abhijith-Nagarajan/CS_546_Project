```python
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
import numpy as np

DB_FAISS_PATH = "db_faiss_research_546"

def create_vector_db(documents):
    # Split documents into chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=550)
    split_docs = text_splitter.split_documents(documents)

    print(f"Total chunks: {len(split_docs)}")

    # Extract texts from split documents
    doc_texts = [doc.page_content for doc in split_docs]
    print(f"Sample text chunk: {doc_texts[0] if doc_texts else 'No text available'}")

    # Initialize embeddings
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2",
        model_kwargs={'device': 'cpu'}
    )

    # Generate embeddings
    print("Generating embeddings...")
    try:
        embedding_vectors = embeddings.embed_documents(doc_texts)
    except Exception as e:
        print(f"Error generating embeddings: {e}")
        return

    # Debug the embeddings
    print(f"Embeddings generated successfully. Type: {type(embedding_vectors)}")
    if not isinstance(embedding_vectors, np.ndarray):
        print("Converting embeddings to NumPy array...")
        embedding_vectors = np.array(embedding_vectors, dtype='float32')

    # Check the shape of the embeddings
    print(f"Embeddings shape: {embedding_vectors.shape}")

    # Ensure embeddings are a 2D NumPy array
    if len(embedding_vectors.shape) != 2:
        raise ValueError("Embedding vectors must be a 2D array.")
    if embedding_vectors.dtype != np.float32:
        embedding_vectors = embedding_vectors.astype('float32')

    # Create the FAISS index
    try:
        db = FAISS.from_documents(split_docs, embeddings)
        db.save_local(DB_FAISS_PATH)
        print(f"FAISS database saved to {DB_FAISS_PATH}")
    except Exception as e:
        print(f"Error creating or saving FAISS index: {e}")

# Example documents (replace 'text' with your input)

create_vector_db(text[:2])
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from collections import defaultdict
import torch
from torch import nn


from sentence_transformers import SentenceTransformer

class AttentionAggregator(nn.Module):
    def __init__(self, embedding_dim):
        random_state=42
        super(AttentionAggregator, self).__init__()
        self.attention_weights = nn.Linear(embedding_dim, 1)

    def forward(self, section_embeddings):
        weights = self.attention_weights(section_embeddings)
        attention_scores = torch.softmax(weights, dim=0)
        aggregated_embedding = torch.sum(section_embeddings, dim=0)
        return aggregated_embedding

def generate_embeddings(all_papers):
    model = Matryoshka(matryoshka_dim=384, adaptor=True)
    doc_emb = []
    
    for i in range(len(all_papers)):
        section_embeddings = {k: model.encode([all_papers[i][k]]) for k in all_papers[i].keys()}
        section_embeddings_tensor = torch.tensor(list(section_embeddings.values()))  # Shape: (num_sections, embedding_dim)
        section_embeddings_tensor = section_embeddings_tensor.squeeze(1)
        aggregator = AttentionAggregator(embedding_dim=section_embeddings_tensor.shape[1])
        document_embedding = aggregator(section_embeddings_tensor)
        doc_emb.append(document_embedding)

    return doc_emb


def process_embeddings(embeddings_list):
    """
    Convert a list of embeddings into a single numpy array.

    Args:
        embeddings_list (list): List of embeddings (numpy arrays or tensors)

    Returns:
        numpy.ndarray: Array of shape (num_samples, embedding_dim)
    """
    # Ensure all embeddings are numpy arrays
    embeddings_array = np.vstack([
        emb.detach().cpu().numpy() if isinstance(emb, torch.Tensor) else np.array(emb)
        for emb in embeddings_list
    ])
    print(f"Processed Embeddings Shape: {embeddings_array.shape}")
    return embeddings_array


def cluster_papers(embeddings, num_clusters=5):
    """
    Clusters the input embeddings into specified number of clusters using KMeans.

    Args:
        embeddings (numpy.ndarray): Shape (num_samples, embedding_dim)
        num_clusters (int): Number of clusters to form

    Returns:
        kmeans (KMeans): Trained KMeans model
        labels (numpy.ndarray): Cluster assignments for each paper
        silhouette (float): Silhouette score indicating clustering quality
    """
    print(f"Clustering {embeddings.shape[0]} papers into {num_clusters} clusters...")


    kmeans = KMeans(n_clusters=num_clusters, init="k-means++", random_state=42, n_init=5)
    labels = kmeans.fit_predict(embeddings)

    
    silhouette = silhouette_score(embeddings, labels)
    print(f"Silhouette Score: {silhouette:.4f} (Closer to 1 indicates better clustering)")

    print("Cluster Centers Shape:", kmeans.cluster_centers_.shape)

    return kmeans, labels, silhouette

embedings_list = generate_embeddings(all_papers)
embeddings_array = process_embeddings(embeddings_list)
kmeans_model, cluster_labels, silhouette = cluster_papers(embeddings_array, num_clusters=5)


print("\nCluster Assignments for Papers:")
for i, label in enumerate(cluster_labels):
    print(f"Paper {i+1}: Cluster {label}")

clusters = defaultdict(list)
for idx, label in enumerate(cluster_labels):
    clusters[label].append(idx + 1)

print("\nPapers grouped by clusters:")
for cluster, papers in clusters.items():
    print(f"Cluster {cluster}: Papers {papers}")
```

```python
import re
import numpy as np
from transformers import pipeline
from sklearn.preprocessing import MinMaxScaler
from sentence_transformers import SentenceTransformer, util
import networkx as nx

# Initialize Hugging Face pipeline and sentence transformer model
def initialize_models():
    sentence_model = SentenceTransformer("all-MiniLM-L6-v2")
    return sentence_model

sentence_model = initialize_models()
```

```python
def extract_known_entities(query, graph):
    """Extract entities from the query using graph nodes."""
    return [node for node in graph.nodes if re.search(rf'\b{node}\b', query, re.IGNORECASE)]
```

```python
def compute_relevance_score(graph, entity):
    """Assign relevance scores to relationships based on edge weights."""
    relevance_scores = {}
    for neighbor in graph.neighbors(entity):
        edge_data = graph.get_edge_data(entity, neighbor)
        edge_weight = edge_data.get("weight", 1)
        relevance_scores[neighbor] = edge_weight
    return relevance_scores
```

```python
def expand_query_with_graph(graph, entities, synonym_model, max_hops=2):
    """
    Expand query terms using graph traversal, relevance scoring, and synonym similarity.
    """
    expanded_terms = {}
    for entity in entities:
        if entity in graph:
            related_terms = {}
            # Graph traversal
            for neighbor, path_length in nx.single_source_shortest_path_length(graph, entity, cutoff=max_hops).items():
                if neighbor != entity:
                    relevance_scores = compute_relevance_score(graph, neighbor)
                    total_score = sum(relevance_scores.values()) / (path_length + 1)  # Normalize by path length
                    related_terms[neighbor] = total_score

            # Synonym expansion
            synonyms = synonym_model.encode([entity] + list(related_terms.keys()), convert_to_tensor=True)
            similarity_scores = util.cos_sim(synonyms[0], synonyms[1:]).squeeze()
            synonym_terms = {list(related_terms.keys())[i]: similarity_scores[i].item() for i in range(len(similarity_scores))}

            # Combine and normalize scores
            related_terms.update(synonym_terms)
            scaler = MinMaxScaler()
            scores = np.array(list(related_terms.values())).reshape(-1, 1)
            normalized_scores = scaler.fit_transform(scores).flatten()
            expanded_terms[entity] = {term: normalized_scores[i] for i, term in enumerate(related_terms.keys())}
        else:
            expanded_terms[entity] = {}
    return expanded_terms
```

```python
def filter_redundancy_and_clustering(expanded_terms, sentence_model, min_community_size=1, add_relation_text=True):
    """
    Filter redundant terms and perform clustering using sentence embeddings.

    Parameters:
    - expanded_terms (dict): Dictionary of entities with their related terms {entity: {term: score}}.
    - sentence_model: SentenceTransformer model for generating embeddings.
    - min_community_size (int): Minimum size for a cluster. Default is 1.
    - add_relation_text (bool): Whether to format sentences with "is related to". Default is True.

    Returns:
    - dict: A dictionary of entities with their clustered, unique related terms.
            Format: {entity: {cluster_index: [related terms]}}
    """
    clustered_terms = {}

    for entity, terms in expanded_terms.items():
        # Prepare sentences for embedding
        related_terms = list(terms.keys())
        if add_relation_text:
            sentences = [f"{entity} is related to {term}" for term in related_terms]
        else:
            sentences = [f"{entity} is related to {term}" for term in related_terms]

        # Generate sentence embeddings
        embeddings = sentence_model.encode(sentences, convert_to_tensor=True)

        # Perform clustering using community detection
        clusters = util.community_detection(embeddings, min_community_size=min_community_size)

        # Map clusters back to terms
        clustered_terms[entity] = {
            f"Cluster {i+1}": [related_terms[idx] for idx in cluster]
            for i, cluster in enumerate(clusters)
        }

    return clustered_terms

```

```python
def generate_llm_prompt(query, clustered_terms):
    """
    Combine the input query and clustered terms into a single, coherent context for LLM input.

    Parameters:
    - query (str): The original query, which can be any medical-related text.
    - clustered_terms (dict): A dictionary of entities with their corresponding clustered terms.

    Returns:
    - str: A combined input text suitable for LLM processing.
    """
    # Collect sentences from clustered terms
    context_sentences = []
    for entity, clusters in clustered_terms.items():
        for cluster_sentences in clusters.values():
            context_sentences.extend(cluster_sentences)

    # Combine query and context
    context = " ".join(context_sentences)
    combined_prompt = (
        f"query: \"{query}\" context: {context}"
    )

    return combined_prompt
```
